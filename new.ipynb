```python
from src.lib.helpers import *
def sentDist(span1,span2):
    doc = span1.doc
    s1 = span1.sent
    s2 = span2.sent
    i1,i2 = -10,500
    s = [x for x in doc.sents]
    for i, sent in enumerate(s):
        if intersectSpanSpan(s1,sent):
            i1=i

        if intersectSpanSpan(s2,sent):
            i2=i
            
    return abs(i1-i2)
```
